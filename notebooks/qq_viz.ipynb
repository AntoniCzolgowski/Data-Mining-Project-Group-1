```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

master = pd.read_csv("../data/processed/master_dataset.csv")
```

```python
#QQ Plot — Median Age
fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(master['median_age'], dist="norm", plot=ax)
ax.set_title("QQ Plot: Median Age", fontsize=14)
ax.get_lines()[0].set(markerfacecolor='steelblue', markeredgecolor='steelblue', markersize=4)
ax.get_lines()[1].set(color='red', linewidth=1.5)
plt.tight_layout()
plt.savefig("../EDAVisualizations/qq_median_age.png", dpi=150)
plt.show()

print(f"Skewness: {master['median_age'].skew():.3f}")
print(f"Kurtosis: {master['median_age'].kurtosis():.3f}")
print(f"Shapiro-Wilk p-value: {stats.shapiro(master['median_age'])[1]:.4f}")
```

```python
# QQ Plot — Democratic Vote Margin
fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(master['dem_margin'], dist="norm", plot=ax)
ax.set_title("QQ Plot: Democratic Vote Margin", fontsize=14)
ax.get_lines()[0].set(markerfacecolor='steelblue', markeredgecolor='steelblue', markersize=4)
ax.get_lines()[1].set(color='red', linewidth=1.5)
plt.tight_layout()
plt.savefig("../EDAVisualizations/qq_dem_margin.png", dpi=150)
plt.show()

print(f"Skewness: {master['dem_margin'].skew():.3f}")
print(f"Kurtosis: {master['dem_margin'].kurtosis():.3f}")
print(f"Shapiro-Wilk p-value: {stats.shapiro(master['dem_margin'])[1]:.4f}")
```

```python
# QQ Plot — Population Density (Raw vs Log-Transformed)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

stats.probplot(master['population_density'], dist="norm", plot=axes[0])
axes[0].set_title("QQ Plot: Population Density (Raw)", fontsize=13)
axes[0].get_lines()[0].set(markerfacecolor='steelblue', markeredgecolor='steelblue', markersize=4)
axes[0].get_lines()[1].set(color='red', linewidth=1.5)

log_density = np.log1p(master['population_density'])
stats.probplot(log_density, dist="norm", plot=axes[1])
axes[1].set_title("QQ Plot: Population Density (Log-Transformed)", fontsize=13)
axes[1].get_lines()[0].set(markerfacecolor='steelblue', markeredgecolor='steelblue', markersize=4)
axes[1].get_lines()[1].set(color='red', linewidth=1.5)

plt.tight_layout()
plt.savefig("../EDAVisualizations/qq_population_density.png", dpi=150)
plt.show()

print(f"Raw  — Skewness: {master['population_density'].skew():.3f}, Kurtosis: {master['population_density'].kurtosis():.3f}")
print(f"Log  — Skewness: {log_density.skew():.3f}, Kurtosis: {log_density.kurtosis():.3f}")
```
